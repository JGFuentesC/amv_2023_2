## Reducción de dimensionalidad

```python
sc = MinMaxScaler()
mds = MDS(n_components=3,n_jobs=-1)
sc.fit(X[varc])
Xm = pd.DataFrame(mds.fit_transform(sc.transform(X[varc])), columns=['d1', 'd2','d3'])
```

```python
Xm[cat] = X[cat]
```

```python
Xm.iplot(kind='scatter3d', x='d1', y='d2',z='d3',
         mode='markers', size=5, categories=cat)
```
